```python
import numpy as np

tokens = ["What", "is", "the", "system", "of", "Amazon"]
seq_len = len(tokens)
embedding_dim = 8



np.random.seed(42)
embeddings = np.random.randn(seq_len, embedding_dim)


W_Q = np.random.randn(embedding_dim, embedding_dim)
W_K = np.random.randn(embedding_dim, embedding_dim)
W_V = np.random.randn(embedding_dim, embedding_dim)



Q = np.dot(embeddings, W_Q)
K = np.dot(embeddings, W_K)
V = np.dot(embeddings, W_V)



scores = np.dot(Q, K.T)


dk = embedding_dim
scaled_scores = scores / np.sqrt(dk)


exp_scores = np.exp(scaled_scores - np.max(scaled_scores, axis=1, keepdims=True))
attention_weights = exp_scores / np.sum(exp_scores, axis=1, keepdims=True)


output = np.dot(attention_weights, V)


print("Attention Weights:\n", attention_weights)
print("\nOutput:\n", output)
```
